### From the above result there is only one country so we need to delete Country and Postal code columns...

```python
sns.heatmap(data.corr(),annot = True, cmap = 'Reds')
```

For quantitative attributes, their linear relationships and their strengths can be compared using a correlation heatmap.
   * Discount and Profit have a weak negative linear correlation. When Discount increase by 1%, Profit will decrease by 0.22 units..
   * Discount and Sales have little to negative linear correlation.

```python
df_numeric = data.select_dtypes(include='number')
df_numeric.plot(kind = 'kde',subplots = True, layout = (2,2), sharex = False, figsize = (15,10))
plt.grid(True)
plt.show()
```

```python
sns.pairplot(data)
plt.show()
```

```python
x = data['Segment']
y = data['Sales']
plt.bar(x,y,)
plt.title('Segments vs Sales')
plt.xlabel('Segments')
plt.ylabel('Sales')
plt.show()
```

### Demonstrate the contribution of Various Segments in the Sales..

```python
plt.figure(figsize =(8,8))
data['Segment'].value_counts().plot(kind='pie',title='Various Segments',legend=True,autopct='%1.1f%%')
```

### Compare the Sales and Profit in various regions of the country.

```python
plt.figure(figsize =(8,8))
data.groupby('Region')['Profit'].sum().plot(kind = 'pie',autopct='%1.1f%%',explode=(0, 0, 0,0.1))
plt.show()
```

```python
data.groupby('Region')['Sales','Profit'].sum().plot.bar()
```

We can observe that:
       * West region of the United States has the more sales and profit..

```python
plt.figure(figsize =(20,10))
data.groupby('Sub-Category')['Sales','Profit'].sum().sort_values(by = ['Profit'] , ascending = True).plot.bar()
plt.ylabel('Sales')
```

```python
data[data['Category']=='Furniture'].groupby('Sub-Category')['Sales','Profit'].sum().plot.bar()
```

Above graph shows the Furniture retails where chairs have gained the higher sales and profit..

```python
data.groupby('State')['Quantity'].sum().sort_values(ascending = False).plot.bar(figsize = (15, 5))
plt.title('Total State-Wise Deals')
plt.ylabel('Total Number of Quantities')
plt.xlabel('American States')
plt.show()
```

```python
plt.figure(figsize =(20,10))
data.groupby('State')['Profit'].sum().sort_values(ascending = False).plot(kind = 'bar')
plt.title('States vs Profit' , fontsize = 20)
plt.grid(True)
plt.show()
```

* California has gained the heighest profit whereas the country should focus more on states which are in the right side.. 

```python
# Sales vs Profit
sns.scatterplot(x= 'Sales', y ='Profit', data =data)
plt.grid(True)
plt.show()
```

```python
sales_discount = pd.pivot_table(data = data, index = 'Region', columns = 'Discount', values = 'Sales', aggfunc = 'sum')
sns.heatmap(sales_discount, cmap = 'Reds')
plt.show()
```

It is showing that there is not much sales after reaching 20% discount..

```python
sales_sub_discount = pd.pivot_table(data = data, index = 'Sub-Category', columns = 'Discount', values = 'Sales', aggfunc = 'sum')
sns.heatmap(sales_sub_discount, cmap = 'Greens')
plt.show()
```

```python
data.groupby('Sub-Category')['Profit'].sum().sort_values(ascending =False).plot.bar(color='g')
plt.ylabel('Profit')
```

```python
 plt.figure(figsize = (8, 5))
# profit/loss by Discount level
sns.lineplot('Discount', 'Profit', data = data, color = 'Brown', label = 'Discount Level')
plt.ylabel('Profit')
plt.title('Profit/Loss by comparing Discount Level', fontsize = 20)
plt.show()
```

## Conclusion

* The discount level should be lowered to 10% to increase the profit.
* The sales should be focused equally on all states in order to reach consumer satisfactory levels.
* The worst losses occured when discount approximates 50%. This may indicate end-of-season sales, and sales for older designs or defect in their products.
* Superstore is most profitable when the discount level reaches lower than 10%. This allows less occurences of lowered prices after decreasing the discount level, and thus more profits.
* We need to reduce the sales of furniture as it is giving less profit for more no. of sales.
